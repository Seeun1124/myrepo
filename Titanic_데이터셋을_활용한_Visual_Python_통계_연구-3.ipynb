# 1. 확률 분포 (Prob Distribution)
- 확률 분포에 해당하는 난수를 생성하여 데이터를 만듦으로써 현실적으로 특정한 구하기 힘든 확률 분포를 가정한 데이터를 쉽게 생성 가능
-  데이터 분석을 통해 생성한 모델의 성능 평가 및 검증 가능

## *이산확률분포(Discrete probability distribution)*
 : 확률 변수가 가질 수 있는 값이 유한하거나 셀 수 있는 무한개인 확률 변수의 확률을 설명하며, 각 값에 대한 확률을 할당

### 1) 베르누이 분포 (Bernoulli)  
: 실험에서 나타날 수 있는 결과가 성공 또는 실패 두 가지로 분류할 수 있을 때 사용하는 분포

`-` 성공 확률(p)이 0.6인 베르누이 시행 결과가 성공이면 1, 실패면 0의 값을 가지는 실험을 1000번 시행한 결과

```python
# Visual Python: Prob. Distribution
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.bernoulli(p=0.6)

# Generate random numbers (Bernoulli)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Bernoulli)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.countplot(x=samples)
    plt.title('Generate random numbers: Bernoulli')
    plt.xlabel('$x$')
    plt.show()
```

### 2) 이항 분포 (Binomial)
: 서로 독립인 베르누이 시행의 합으로 성공확률이 p인 시행을 n번 반복할 때 성공 횟수의 분포


`-` 1000 개의 난수를 생성하여 성공확률(p)이 0.6인 시행을 10(n)번 반복할 때 성공횟수 결과

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.binom(n=10, p=0.6)

# Generate random numbers (Binomial)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Binomial)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.countplot(x=samples)
    plt.title('Generate random numbers: Binomial')
    plt.xlabel('$x$')
    plt.show()
```

###  3) 다항 분포(Multinomial)
: 한 번에 상호배반인 여러 개의 베르누이 시행의 결과를 나타냄

`-` 1000개의 난수를 생성하여 성공확률이 0.4이고, 실패확률이 0.6인 실험을 10번(n) 반복시행한 결과

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.multinomial(n=10, p=[0.4, 0.6])

# Generate random numbers (Multinomial)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Multinomial)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    for i in range(0, samples.shape[1]):
        plt.subplot(2, 2, i+1)
        plt.title('$x$=' + str(i))
        sns.countplot(x=[ x[i] for x in samples ])
    plt.suptitle('Generate random numbers: Multinomial')
    plt.tight_layout()
    plt.show()
```

## *연속확률분포(Continuous probablity distribution)*
- 확률 변수가 연속적인 값을 가지는 가지는 경우에 대한 확률 분포
- 각 값에 확률을 할당하는 것이 아닌 확률 변수가 특정 구간에 속할 확률을 계산

### 1) 균등 분포 (Uniform)
- 특정 구간의 확률 변수가 동일한 확률로 발생하는 경우를 나타내는 분포

`-` 구간 [0,1]에서 임의로 수를 선택하는 것을 10000번 반복했을 때의 분포

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.uniform()

# Generate random numbers (Uniform)
from IPython.display import display
samples = _rv.rvs(size=10000)
display(samples)

# Sample distribution (Uniform)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(samples, stat='density', kde=True)
    plt.title('Generate random numbers: Uniform')
    plt.xlabel('$x$')
    plt.show()
```

- **구간 내에서 확률이 동일하므로 그래프가 직사각형 모양의 평평한 형태를 가짐**

### 2) 정규 분포 (Normal)
- 통계적 방법에서 가장 많이 사용되는 분포
- 중심극한정리에 따르면 표본의 크기가 충분히 크다면, 표본의 평균은 정규 분포에 근사함

`-` 평균(loc)이 0이고, 표준편차(scale)가 1인 난수를 10000개 생성한 결과

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.norm(loc=0, scale=1)

# Generate random numbers (Normal)
from IPython.display import display
samples = _rv.rvs(size=10000)
display(samples)

# Sample distribution (Normal)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(samples, stat='density', kde=True)
    plt.title('Generate random numbers: Normal')
    plt.xlabel('$x$')
    plt.show()
```

- **종 모양의 대칭적인 분포가 나타남**
- **평균을 중심으로 좌우 대칭**
- **평균 주위에 데이터가 많고, 평균에서 멀어질수록 데이터가 적어짐**

### 3) F-분포 (F)  
- 두 개 이상의 표본 분산을 비교하는 데 사용되는 확률 분포
- 주로 ANOVA에서 사용되는 분포

`*` 자유도 파라미터
- Dfn: 분모 자유도이며, 오차(또는 그룹 내의 변동)을 나타냄. 일반적으로 데이터의 개수에서 그룹 수를 뺀 값임.    
- Dfd: 분자 자유도이며, 그룹 간의 변동을 나타냄. 보통 그룹 수에서 1을 뺀 값을 사용하는 것이 일반적     
- 자유도가 높을수록 분석에서 사용되는 통계량의 분포가 더 정확하게 근사됨

> 이를 Titanic 데이터에 적용하여 부모 자식수(Parch)에 따라 그룹을 나눈다고 가정하면 그룹 수는 7임.  
따라서 분모 자유도는 891(데이터의 총 개수) - 7(그룹 수) = 884이며, 분자 자유도는 7(그룹 수) - 1 = 6임.

`-` Dfn이 884이고, Dfd가 6인 난수 10000개의 분포

```python
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.f(dfn=884, dfd=6)

# Generate random numbers (F)
from IPython.display import display
samples = _rv.rvs(size=10000)
display(samples)

# Sample distribution (F)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(samples, stat='density', kde=True)
    plt.title('Generate random numbers: F')
    plt.xlabel('$x$')
    plt.show()
```

- **종 모양이며, 오른쪽으로 꼬리가 긴 형태**

# 3. 정규성 검정 (Normality test)
- 많은 통계 기법들이 데이터가 정규 분포를 따른다는 가정에 기초하므로 적절한 분석 방법을 선택하는데 활용 가능
- 정규 분포를 따르지 않는다는 결과가 도출되었을 때, 데이터를 표준화하여 통계적 추론의 정확성을 높일 수 있음
- 통계량과 시각화 결과를 통해 데이터가 정규 분포를 따르는지 직관적으로 확인 가능
- Q-Q Plot이 가장 일반적으로 사용되는 시각화 방법

## `*` 정규성 검정의 가설
> 귀무가설: 데이터는 정규 분포를 따른다.  
대립가설: 데이터는 정규 분포를 따르지 않는다.



## `*` Q-Q Plot 해석 방법
- 빨간 직선: 이론적 분포
- 파란 점: 데이터의 값
- x축은 이론적 분위수, y축은 데이터의 실제 값을 나타냄
> 1)  
2)

1) 연속 변수인 'Age'가 정규성을 확인하기 위해 Shapiro-Wilk test 시행

```python
# Visual Python: Normality test
# Shapiro-Wilk test
vp_df = titanic.dropna().copy()

# Normality test (Shapiro-Wilk)
from IPython.display import display, Markdown
from scipy import stats
_res = stats.shapiro(vp_df['Age'])
display(Markdown('### Normality test (Shapiro-Wilk)'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},index=['Normality test (Shapiro-Wilk)']))

# Charts
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)

    plt.subplot(2,2,1)
    stats.probplot(vp_df['Age'], plot=plt)
    plt.title('Q-Q Plot')

    plt.tight_layout()
    plt.show()
```

`-` **해석**  
 - 통계량이 0.9 이상이므로 정규 분포를 따름
 - p-value가 유의수준 0.05보다 크기 때문에 'Age' 변수는 정규 분포를 따른다는 귀무가설을 기각하지 못 하므로 정규 분포를 따름  
 - 그래프와 대각선의 형태가 유사하여 정규 분포와 거의 일치

# 4. 등분산 검정 (Equal Var. test)
- 두 개 이상의 그룹 간에 분산이 동일한지 확인하기 위해 사용됨.   

## `*` 등분산 검정의 가설
> 귀무가설: 두 집단의 분산이 같다.  
대립가설: 두 집단의 분산이 다르다.



`-` 'Survived' 변수에 따른 그룹 간 'Age' 변수의 Levene test 결과
(Variable이 'Age', Factor가 'Survived')

```python
# Visual Python: Equal Var. test
# Levene test
vp_df = titanic.dropna().copy()
_df = pd.DataFrame()
for k, v in  dict(list(vp_df.groupby('Survived')['Age'])).items():
    _df_t = v.reset_index(drop=True)
    _df_t.name = k
    _df = pd.concat([_df, _df_t], axis=1)

# Variance
from IPython.display import display, Markdown
from scipy import stats
_dfr = _df.var().to_frame()
_dfr.columns = ['Variance']
display(_dfr)

# Levene test
_lst = []
_df.apply(lambda x: _lst.append(x.dropna()))
_res = stats.levene(*_lst, center='mean')
display(Markdown('### Levene test'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},
                     index=['Equal Variance test (Levene)']))

# Histogram
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(_df, stat='density', kde=True)
    plt.title('Histogram')
    plt.show()
```

`-` **해석**
- p-value가 유의 수준보다 높기 때문에 귀무가설을 기각하지 못 한다. 따라서 생존한 그룹과 생존하지 못 한 그룹의 나이의 분산이 같다.
