# 1. 확률 분포 (Prob Distribution)
- 확률 분포에 해당하는 난수를 생성하여 데이터를 만듦으로써 특정한 현실적으로 구하기 힘든 확률 분포를 가정한 데이터를 쉽게 생성 가능

## 이산확률분포(Discrete probability distribution)
- 확률 변수가 가질 수 있는 값이 유한하거나 셀 수 있는 무한개인 확률 변수의 확률을 설명하며, 각 값에 대한 확률을 할당

### 베르누이 분포 (Bernoulli)
- 실험에서 나타날 수 있는 결과가 성공 또는 실패 두 가지로 분류할 수 있을 때 사용

`-` 성공 확률(p)이 0.6인 베르누이 시행 결과가 성공이면 1, 실패면 0의 값을 가지는 실험을 1000번 시행한 결과

```
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.bernoulli(p=0.6)

# Generate random numbers (Bernoulli)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Bernoulli)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.countplot(x=samples)
    plt.title('Generate random numbers: Bernoulli')
    plt.xlabel('$x$')
    plt.show()
```

`-` 해석: 성공이 실패보다 더 많음

### 이항 분포 (Binomial)
- 서로 독립인 베르누이 시행의 합으로 성공확률이 p인 시행을 n번 반복할 때 성공 횟수의 분포


`-` 1000 개의 난수를 시행하여 성공확률(p)이 0.6이며 10(n)번 반복할 때 성공횟수 결과

```
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.binom(n=10, p=0.6)

# Generate random numbers (Binomial)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Binomial)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.countplot(x=samples)
    plt.title('Generate random numbers: Binomial')
    plt.xlabel('$x$')
    plt.show()
```

### 다항 분포(Multinomial)
- 한 번에 상호배반인 여러 개의 베르누이 시행의 결과를 나타냄

`-` 1000개의 난수를 생성하여 성공확률이 0.4이고, 실패확률이 0.6인 실험을 10번(n) 반복시행한 결과

```
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.multinomial(n=10, p=[0.4, 0.6])

# Generate random numbers (Multinomial)
from IPython.display import display
samples = _rv.rvs(size=1000)
display(samples)

# Sample distribution (Multinomial)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    for i in range(0, samples.shape[1]):
        plt.subplot(2, 2, i+1)
        plt.title('$x$=' + str(i))
        sns.countplot(x=[ x[i] for x in samples ])
    plt.suptitle('Generate random numbers: Multinomial')
    plt.tight_layout()
    plt.show()
```

### F-분포 (F): ANOVA에서 사용되는 분포

Dfn은 분모 자유도이며, 오차(또는 그룹 내의 변동)을 나타냄. 일반적으로 데이터 포인트의 총 개수에서 그룹 수를 뺀 값임.    
Dfd는 분자 자유도이며, 그룹 간의 변동을 나타냄. 보통 그룹 수에서 1을 뺀 값을 사용하는 것이 일반적.

이를 Titanic 데이터에 적용하여 선실 등급(Pclass)에 따라 그룹을 나눈다고 가정하면 그룹 수는 3임.  
따라서 분모 자유도는 891(데이터 포인트의 총 개수) - 3(그룹 수) = 888이며, 분자 자유도는 3(그룹 수) - 1 = 2임.

```
# Visual Python: Prob. Distribution
from scipy import stats

_rv = stats.f(dfn=50, dfd=1)

# Generate random numbers (F)
from IPython.display import display
samples = _rv.rvs(size=10000)
display(samples)

# Sample distribution (F)
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(samples, stat='density', kde=True)
    plt.title('Generate random numbers: F')
    plt.xlabel('$x$')
    plt.show()
```

# 3. 정규성 검정 (Normality test)
`-` (목적) 많은 통계 기법들이 데이터가 정규 분포를 따른다는 가정에 기초하므로 적절한 분석 방법을 선택하는데 활용  
`-` (목적2) 정규 분포를 따르지 않는다는 결과가 도출되었을 때, 데이터를 표준화하여 통계적 추론의 정확성을 높일 수 있음

- 연속 변수인 'Age'가 정규 분포를 따르는지 여부를 확인하기 위해 정규성 검정인 Shapiro-Wilk test를 시행 (정규성을 따르는 경우)

```
# Visual Python: Normality test
# Shapiro-Wilk test
vp_df = titanic.dropna().copy()

# Normality test (Shapiro-Wilk)
from IPython.display import display, Markdown
from scipy import stats
_res = stats.shapiro(vp_df['Age'])
display(Markdown('### Normality test (Shapiro-Wilk)'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},index=['Normality test (Shapiro-Wilk)']))

# Charts
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)

    plt.subplot(2,2,1)
    stats.probplot(vp_df['Age'], plot=plt)
    plt.title('Q-Q Plot')

    plt.tight_layout()
    plt.show()
```

`*` Q-Q Plot:

`-` 귀무가설: 'Age' 변수는 정규 분포를 따른다.  
`-` 대립가설: 'Age' 변수는 정규 분포를 따르지 않는다.    
`-` **검정 결과, p-value가 유의수준 0.05보다 크기 때문에 'Age' 변수는 정규 분포를 따른다는 귀무가설을 기각하지 못 한다. 따라서 'Age' 변수는 정규 분포를 따른다.**

---

- 연속 변수인 'Fare'가 정규 분포를 따르는지 여부를 확인하기 위해 정규성 검정인 Shapiro-Wilk test를 시행 (정규성을 따르지 않는 경우)

```
# Visual Python: Normality test
# Shapiro-Wilk test
vp_df = titanic.dropna().copy()

# Normality test (Shapiro-Wilk)
from IPython.display import display, Markdown
from scipy import stats
_res = stats.shapiro(vp_df['Fare'])
display(Markdown('### Normality test (Shapiro-Wilk)'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},index=['Normality test (Shapiro-Wilk)']))

# Charts
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)

    plt.subplot(2,2,1)
    stats.probplot(vp_df['Fare'], plot=plt)
    plt.title('Q-Q Plot')

    plt.tight_layout()
    plt.show()
```

`-` 귀무가설: 'Fare' 변수는 정규 분포를 따른다.  
`-` 대립가설: 'Fare' 변수는 정규 분포를 따르지 않는다.    
`-` **검정 결과, p-value가 유의수준 0.05보다 작기 때문에 'Fare' 변수는 정규 분포를 따른다는 귀무가설을 기각한다. 따라서 'Fare' 변수는 정규 분포를 따르지 않는다.**

# 4. 등분산 검정 (Equal Var. test)
`-` 왜 시행하는지?: 두 개 이상의 그룹 간에 분산이 동일한지 확인하기 위해 사용됨.   
`-` 이 검정의 결과를 어떻게 해석하는지?

- 생존 여부에 따른 그룹 간 나이 변수의 등분산 검정 수행

```
# Visual Python: Equal Var. test
# Levene test
vp_df = titanic.dropna().copy()
_df = pd.DataFrame()
for k, v in  dict(list(vp_df.groupby('Survived')['Age'])).items():
    _df_t = v.reset_index(drop=True)
    _df_t.name = k
    _df = pd.concat([_df, _df_t], axis=1)

# Variance
from IPython.display import display, Markdown
from scipy import stats
_dfr = _df.var().to_frame()
_dfr.columns = ['Variance']
display(_dfr)

# Levene test
_lst = []
_df.apply(lambda x: _lst.append(x.dropna()))
_res = stats.levene(*_lst, center='mean')
display(Markdown('### Levene test'))
display(pd.DataFrame(data={'Statistic':_res.statistic,'p-value':_res.pvalue},
                     index=['Equal Variance test (Levene)']))

# Histogram
import seaborn as sns
import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=Warning)
    sns.histplot(_df, stat='density', kde=True)
    plt.title('Histogram')
    plt.show()
```

`-` 귀무가설: 생존한 그룹과 생존하지 못 한 그룹의 나이의 분산이 같다.  
`-` 대립가설: 생존한 그룹과 생존하지 못 한 그룹의 나이의 분산이 다르다.  
`-` **검정 결과, p-value가 유의 수준보다 높기 때문에 귀무가설을 기각하지 못 한다. 따라서 생존한 그룹과 생존하지 못 한 그룹의 나이의 분산이 같다.**
